**Import Library Pandas, Matplotlib, dan Seaborn**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(color_codes=True)

import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("iris.csv")
df.head()
```

```python
# Membuat Scatter plot dari fitur SepalLengthCm dan SepalWidthCm
df.plot(kind="scatter", x="SepalLengthCm", y="SepalWidthCm")
```

```python
# Atau dengan library Seaborn
sns.jointplot(x="SepalLengthCm", y="SepalWidthCm", data=df, size=5)
```

```python
# Salah satu informasi yang hilang dalam plot di atas adalah jenis tanaman(Species)
# Gunakan FacetGrid Seaborn untuk mewarnai sebaran Species
sns.FacetGrid(df, hue="Species", palette="husl") \
   .map(plt.scatter, "SepalLengthCm", "SepalWidthCm") \
   .add_legend()
```

```python
# 3 plot dengan jenis Palette atau warna yang berbeda
sns.FacetGrid(df, hue="Species", palette="hls") \
   .map(plt.scatter, "SepalLengthCm", "SepalWidthCm") \
   .add_legend()

sns.FacetGrid(df, hue="Species", palette="Set2") \
   .map(plt.scatter, "SepalLengthCm", "SepalWidthCm") \
   .add_legend()

sns.FacetGrid(df, hue="Species", palette="cubehelix") \
   .map(plt.scatter, "SepalLengthCm", "SepalWidthCm") \
   .add_legend()
```

```python
# pairplot
# Dibawah ini ada kolom Id yang dihapus karena tidak memiliki korelasi dengan variabellain
sns.pairplot(df.drop("Id", axis=1), hue="Species", palette="husl")
```
